```python
import matplotlib.pyplot as plt

source_counts = df['source'].value_counts()

cmap = ['#8B8EFF', '#FE7070', '#FFE341', '#05B9A0']

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{:.1f}%\n({v:d})'.format(pct, v=val)
    return my_format

plt.figure(figsize=(10, 6))
plt.pie(source_counts, labels=source_counts.index.tolist(), autopct=autopct_format(source_counts), colors=cmap)
plt.title('Distribution of Fungal Data Sources')
plt.ylabel('')
plt.show()
```
