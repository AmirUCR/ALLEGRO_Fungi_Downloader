```python
import pandas as pd

df = pd.read_csv("species_w_classifications.csv")
df
```

```python
subset = df[df['classification'].str.contains('Ascomycota;')].copy()
subset['species_name'] = subset['species_name'].str.replace(' ', '_').str.lower()
subset = subset.reset_index(drop=True)

subset
```

```python
input_species = pd.read_csv('../../ALLEGRO/data/input/fourdbs_input_species.csv')
input_species_subset = input_species[input_species['species_name'].isin(subset['species_name'])].reset_index(drop=True)
input_species_subset
```

```python
from Bio import SeqIO

exclusion = list()

for idx, row in input_species_subset.iterrows():
    path = f'../data/fourdbs_concat/cds_from_gff/{row["cds_file_name"]}'
    l = len(list(SeqIO.parse(path, 'fasta')))

    if l < 2000:
        print(row['cds_file_name'], l)
        exclusion.append(row['species_name'])
```

```python
input_species_subset = input_species_subset[~input_species_subset['species_name'].isin(exclusion)]
input_species_subset.to_csv('ascomycota_input_species.csv', index=False)
```
